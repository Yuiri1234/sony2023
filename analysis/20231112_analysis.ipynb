```python
import numpy as np
import pandas as pd
import os
import sys
import cv2
import matplotlib.pyplot as plt

sys.path.append('../')
```

```python
from configs.config import CONF
```

```python
X_test = np.load(os.path.join(CONF.PATH.DATASET, "all", "X_test.npy"))
X_test = X_test.reshape(-1, 20, 20)
X_test.shape
```

```python
def plot_result(test_folder, X_test, i=0):
    csv_file = test_folder.split("_")[0] + "_result.csv"
    y_test_pred = pd.read_csv(os.path.join(CONF.PATH.OUTPUT, test_folder, csv_file), header=None, index_col=0)
    fig, axes = plt.subplots(10, 10, figsize=(30, 30))
    for m in range(10):
        for n in range(10):
            k = 10*10*i + m*10 + n
            test = X_test[k]
            pred = y_test_pred.iloc[k, 0]
            axes[m, n].imshow(test, cmap='gray', vmin=0, vmax=255)
            axes[m, n].set_title(f"index:{k} label: {pred}")
            axes[m, n].axis('off')
    plt.show()
```

```python
test_folder = "20231112-081035_all_default_resnet"
plot_result(test_folder, X_test, i=0)
```
