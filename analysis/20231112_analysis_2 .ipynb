```python
import numpy as np
import pandas as pd
import os
import sys
import cv2
import matplotlib.pyplot as plt

sys.path.append('../')
```

```python
from configs.config import CONF
```

```python
X_train_original = np.load(os.path.join(CONF.PATH.DATASET, "all", "X.npy"))
X_train_original = X_train_original.reshape(-1, 20, 20)
X_train_noise = np.load(os.path.join(CONF.PATH.DATASET, "noise", "X.npy"))
X_train_noise = X_train_noise.reshape(-1, 20, 20)
X_train_comp = np.load(os.path.join(CONF.PATH.DATASET, "all+three", "X_train_filtered.npy"))
X_train_comp = X_train_comp.reshape(-1, 20, 20)
X_train_comp_noise = np.load(os.path.join(CONF.PATH.DATASET, "noise+three", "X_train_filtered.npy"))
X_train_comp_noise = X_train_comp_noise.reshape(-1, 20, 20)
X_test_original = np.load(os.path.join(CONF.PATH.DATASET, "all", "X_test.npy"))
X_test_original = X_test_original.reshape(-1, 20, 20)
X_train_original.shape, X_train_noise.shape, X_train_comp.shape, X_train_comp_noise.shape, X_test_original.shape
```

```python
def plot_result(X, i=0):
    fig, axes = plt.subplots(10, 10, figsize=(30, 30))
    for m in range(10):
        for n in range(10):
            k = 10*10*i + m*10 + n
            test = X[k]
            axes[m, n].imshow(test, cmap='gray', vmin=0, vmax=255)
            axes[m, n].set_title(f"index:{k}")
            axes[m, n].axis('off')
    plt.show()
```

```python
plot_result(X_train_original, i=0)
```

```python
plot_result(X_train_noise, i=0)
```

```python
plot_result(X_test_original, i=0)
```

```python
plot_result(X_train_comp, i=0)
```

```python
plot_result(X_train_comp_noise, i=0)
```
